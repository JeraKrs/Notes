# 总结

## 操作系统

**中断**：指的是计算机在执行期间，系统内发生任何非寻常的或非预期的急需处理事件，使得 CPU 暂时中断当前正在执行的程序而转去执行相应的事件处理程序，待处理完毕之后又返回原来被中断处继续执行或调度新的进程执行的过程。

## 内存管理

**内存分配算法**：为进程分配内存。分为连续内存分配和非连续内存分配。
* **连续内存分配**：首次适配算法、最优适配算法、最坏适配算法、伙伴算法。
* **非连续内存分配**：分段机制、分页机制、多级页表机制、反向页表。

**页面置换算法**：在地址映射的过程中发生缺页中断的现象时，选择一个页面将其移出内存。页面置换算法分为固定驻留策略和动态驻留策略，规定驻留策略必然导致局部置换，而全局置换必须使用动态驻留策略。
* 固定驻留策略：每个进程在内存中的驻留集大小固定；**局部页面置换算法**：最佳置换算法、先进先出算法、最近最久未使用算法、时钟置换算法、二次机会法、最不常用算法。
* 动态驻留策略：进程在内存中的驻留集大小可变；**全局页面置换算法**：工作集置换算法、可变分配策略。

## 进程和线程

**进程间通信方式**：管道、信号、消息队列、共享内存、信号量、套接字。

**进程和线程的区别**：
* 进程是系统进行资源分配的基本单位，线程是系统执行的基本单位。
* 一个进程包含一个以上的线程，一个线程只属于一个进程。
* 进程间不共享资源，同一个进程内的线程共享进程的资源。
* 多进程比多线程健壮，一个线程死掉说明进程死掉了，但是一个进程死掉对另 一个进程是没有影响的。

**进程的状态**：运行状态，就绪状态，阻塞状态，创建状态，结束状态，就绪挂起，阻塞挂起。

**同步和异步的区别**：同步是阻塞模式，即发送方发出数据之后需要等待接收方响应之后才能继续发送下一个数据包；异步是非阻塞模式，发送方发出数据之后，不用等待接收方响应，而是直接发送下一个数据包。

**线程安全**：多线程运行结果是可预期的，而且与单线程的运行结果一样。

**进程调度算法**：一个进程从运行状态切换到等待状态或被终结时，从就绪态的进程中挑选出一个获得CPU资源。先进先出算法、短进程优先算法、最高响应比优先、轮循调度算法、多级队列反馈、公平共享调度。

**线程池**：在任务到来之前，创建一定数量的线程，放入空闲队列中。这些线程都是处于睡眠状态，即均为启动，不消耗 CPU，而只是占用较小的内存空间。当请求到来之后，缓冲池给这次请求分配一个 空闲线程，把请求传入此线程中运行，进行处理。

**线程池的好处**：减少在创建和销毁线程上所花的时间以及系统资源的开销；避免系统因创建大量线程而导致系统内存被消耗完。

## 死锁

**什么是死锁**：死锁是指两个或两个以上的进程在执行的过程中，因争夺资源而造成的互相等待的现象，如果没有外力的作用无法继续推进。

**死锁产生的四个必要条件**：互斥、持有并等待、无抢占、循环等待。

**死锁的四种解决办法**：死锁预防、死锁避免、死锁检测、死锁恢复。

## 文件系统

**文件别名**：两个或多个文件名关联同一个文件。
* 硬连接：多个文件项指向一个文件，由文件的inode记录有多少个文件项指向该文件。（硬链接不可以跨越文件系统；硬链接不可以对不存在的文件链接；硬链接是不可以对目录进行链接的）
* 软连接：以“快捷方式”指向其他文件，”快捷方式“保存了真正文件的路径。

**删除一个有别名的文件**
* 硬连接中有一个计数器，记录了连接的个数，只有连接个数为零时删除文件。
* 软连接的“快捷方式”则会被为一个悬空指针。
