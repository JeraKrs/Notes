# 第二讲 操作系统的启动、中断、异常和系统调用

## 2.1 操作系统的启动

1. POST 加电自检，寻找显卡和执行BIOS。
2. BIOS（Basic Input Output System）进程位于内存的固定位置，为基本的I/O处理系统，它会从磁盘的默认位置（一般为第一个扇区）加载一个大小为512字节的 Bootloader 程序。
3. Bootloader 程序会将保存在磁盘中的操作系统加载到内存中，并将执行权限交给操作系统。

## 2.2 系统调用、异常、中断

**定义**

* 系统调用：应用程序主动向操作系统发出服务请求。
* 异常：非法指令或其他坏的处理状态，如内存出错。
* 中断：来自不同硬件设备的计时器和网络的中断。

**为什么应用程序不能直接访问外部设备？**

* 内核是被信任的第三方软件。
* 只有内核可以执行特权指令。
* 为了方便应用程序。

**系统调用、异常、中断之间的差异**

* 源头：
    * 中断：来源于外部设备，如键盘、鼠标、网卡、声卡、显卡等。
    * 异常：应用程序意想不到的行为，如除零操作，恶意程序等。
    * 系统调用：应用程序主动请求操作系统提供服务。
* 处理时间：
    * 中断：异步操作。
    * 异常：同步操作。
    * 系统调用：一般为同步操作，而在一些特殊情况下为异步的。
* 响应：
    * 中断：持续的，对应用程序透明。
    * 异常：杀死或者重新执行异常指令。
    * 系统调用：等待和持续。

**中断的处理机制**

* 硬件会设置中断标记，中断标记会指明中断所属类型。
* 软件层面，操作系统会先保存当前被打断的应用程序的处理状态；然后根据中断标记，查找中断表中对应的处理机制并跳转执行；在执行完对应的处理机制后，清空中断标记；最后恢复之前被打断的应用程序的处理状态。

**异常的处理机制**

先保存现场；然后进行异常处理，异常处理有两种方式，一是杀死异常的程序，一是为异常指令提供所需的条件后重新尝试执行；最后恢复程序的执行状态。

**系统调用的处理机制**

操作系统会从用户态切换到内核态，然后执行相应的服务，最后切换回用户态。

* 用户态：应用程序在执行中，CPU所处于的特权级状态。
* 内核态：操作系统在执行中，CPU所处于的特权级状态，可以执行任何一条指令。

应用程序访问主要通过高层次的API接口，而不是直接进行系统调用：
* Win32 API 用于Windows
* POSIX API 用于POSIX - based systems（Unix，Linux，Mac OS）
* JAVA API 用于 JAVA 虚拟机（JVM）

**系统调用和函数调用的区别**

* 系统调用中服务和应用程序不共享堆栈，操作系统有自己的堆栈；需要进行特权级切换。
* 函数调用中函数服务与应用程序共享堆栈。

**开销**

* 建立中断、异常、系统调用号与之对应服务映射关系的初始化开销。
* 建立内核堆栈。
* 验证参数开销。
* 内核态映射到用户态的地址空间。
* 内核态独立的地址空间。
