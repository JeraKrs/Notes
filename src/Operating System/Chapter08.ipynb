# 第八章 调度算法

## 8.1 调度算法背景

内核运行调度程序的条件（满足一条即可）：

* 一个进程从运行状态切换到等待状态。
* 一个进程被终结。

调度算法分为可抢占式和不可抢占式：
* 不可抢占式：调度程序必须等待事件结束。
* 可抢占式：
    * 调度程序在中断被响应后执行。
    * 当前进程从运行切换到就绪或一个进程从等待切换。
    * 当前运行的进程可以被换出。

## 8.2 调度原则

**执行模型**：程序在CPU突发和I/O中交替。

**调度算法的评价指标**：

* CPU使用率：CPU处于忙状态所占时间的百分比。
* 吞吐率：在单位时间内完成的进程数量。
* 周转时间：一个进程从初始化到结束，包括所有等待时间。
* 等待时间：进程在就绪队列中的总时间。
* 响应时间：从一个请求被提交到产生第一次响应所花费的总时间。

**调度算法目标**：减少响应时间、减少平均响应时间的波动、增加吞吐量、减少等待时间。

## 8.3 调度算法

**先来先服务（First Come First Served，FCFS）**：如果进程在执行中阻塞，队列中的下一个会得到CPU。它的优点是简单，缺点是平均等待时间波动较大；花费时间少的任务可能排在花费时间长的任务后面；可能导致I/O和CPU之间的重叠处理。

**短进程优先/最剩余时间优先（Shortest Process Next/Shortest Remaining Time，SPN/SRT）**：对应的是非抢占、抢占式，它会选择一个时间最短的进程。它的优点是平均等待时间最少，缺点是可能导致饥饿；并且需要预知程序的运行时间。

**最高响应比优先（Highest Response Ratio Next，HRRN）**：在SPN调度的基础上改进，不考虑抢占，它关注进程等待了多长时间，防止了无限期推迟。$R = (w + s)/s$，其中w为等待时间（Waiting Time），s为执行时间（Service Time），调度算法会选择R值最高的进程。对比SPN，它解决了饥饿问题，但响应比仍然是需要预估的。

**轮循调度算法（Round Robin，RR）**：将时间分成若干时间片，称作量子或时间片的离散单元，时间片结束时，且换到下一个准备好的进程。它的缺点在于：
* 花销：额外的上下文切换。
* 时间量子太大：等待的时间过长，极限情况下退化成FCFS。
* 时间量子太小：反应迅速，吞吐量由于大量的上下文切换开销收到影响。

**多级队列反馈（Multilevel Feedback Queues，MFQ）**：

* 就绪队列被划分成若干个独立的队列。
* 每个队列拥有自己的调度策略。
* 一个进程可以在不同队列中移动。
* 进程消耗CPU时间越多，优先级降低；进程等待时间越多，优先级增加。

**公平共享调度（Fair Share Sheduling，FSS）**：公平是第一要素。

## 8.4 实时调度

**实时调度系统**：正确性依赖于其时间和功能两方面的一种操作系统。

**性能指标**：时间约束的可预测性。

**主要特征**：

* 强实时系统：需要在要求的时间内完成重要任务。
* 弱实时系统：需要在要求的时间内尽量完成重要任务。

**实时调度算法**：
* 速率单调调度（Rate Monotonie）：优先级在任务执行前确定。
* 最早期限调度（Earliest Deadline First）：deadline越早优先级越高。

## 8.5 多处理器调度与优先级反转

**多处理器**：多个相同的单处理器组成一个多处理器，它的优点是负载共享。

优先级反转的持续时间决于其他不相关任务的不可预测的行为。

解决办法：

* 优先级继承低优先级继承者优先级依赖于他们的共享的资源。
* 对进程所需的资源进行定级，资源的优先级与访问它的所有进程中最高的优先级相同，某低优先级提升至资源优先级的水平。
