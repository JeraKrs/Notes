# 第三讲 内存地址和连续内存的分配

## 3.1 计算机体系结构及内存分层体系

**计算机体系结构**

* CPU：对程序和软件执行的控制。
* 内存：程序的代码和所处理的数据。
* 外设：磁盘、屏幕等。

**对物理内存管理的目标**

* 抽象：逻辑地址空间。
* 保护：不同程序间拥有独立地址空间。
* 共享：访问相同内存。
* 虚拟化：更多的地址空间。

**操作系统管理内存的方法**：程序重定位、分段、分页、虚拟内存、按需分页虚拟内存。


## 3.2 地址空间与地址生成

**地址空间定义**

* 物理地址空间与硬件直接对应的，由硬件管理。
* 逻辑地址空间，一个运行的应用程序拥有的内存范围。

**地址生成**：地址由应用程序、编译器、加载器生成。当 CPU 需要访问某个虚拟地址时，会在MMU上查找逻辑地址的映射表，如果存在映射关系，则返回物理地址；如果不存在，则在内存中的映射表上查询。

**安全检查**：操作系统会确保每个应用程序的有效地址空间，通过记录起始地址和地址长度的方式。CPU 执行某条指令前，会先查询映射表，而这个查询过程会指明地址是否合法。

## 3.3 连续内存分配：内存碎片与分区的动态分配

**内存碎片问题**：空间碎片不能被利用。

* 外部碎片：在分配单元间的未使用内存。
* 内部碎片：在分配单元内的未使用内存。

**操作系统何时需要分配内存**

* 当一个程序准许运行在内存中时。
* 分配一个连续的内存区间给运行的程序以访问数据。

**分配策略**

* 首次适配：为了分配n个字节，使用第一个可用空闲块。优势为简单，易于产生更大的空闲块；劣势为会产生外部碎片，以及具有不确定性。
* 最优适配：为了分配n个字节，使用最小的可用空闲块。优势为在大部分分配为小尺寸时非常有效，而且比较简单；劣势为会产生外部碎片，重分配慢，而且易于产生很多没有用的微小碎片。
* 最坏适配：为了分配n个字节，使用最大的可用空闲块。优势为在大部分分配为中等尺寸时非常有效；劣势为重分配慢，会产生外部碎片，而且易于破碎大的空闲块以致于大分区无法被分配。

内存分配算法：

* 需要按照地址或大小排序空闲块。
* 分配需要寻找一个适合的分区。
* 重分配时需要搜索和合并相邻的空闲块。

## 3.4 压缩式与交换式碎片整理

**压缩式碎片整理**：重置程序内存地址以合并碎片。

**交换式碎片整理**：抢占等待程序并回收它们的内存。
