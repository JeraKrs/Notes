## 线性相关性

对于一组向量 $v_1, v_2, \dots, v_n$ 而言，如果没有任何非零的线性组合可以得到零向量，那么这组向量就是线性无关的（Linear Independence）。如果这组向量组成矩阵 $A$，那么矩阵 $A$ 的零空间中有且仅有 $0$ 向量，此时 $rank(A)=n$，并且没有自由变量；当存在非零向量 $c$ 使得 $Ac=0$，则存在线性相关向量，此时 $rank(A)<n$，并且有自由变量。

```python
%matplotlib inline
from functools import partial
import matplotlib.pyplot as plt
import numpy as np

plt.axhline(y=0, c='gray')
plt.axvline(x=0, c='gray')

ax = plt.gca()
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

arrow_vector = partial(plt.arrow, width=0.05, length_includes_head=True)

arrow_vector(0, 0, 1, 1,  color='black')
arrow_vector(0, 0, 2, 2, color='black')
plt.title('Linear Dependence')
plt.show()
```

```python
plt.axhline(y=0, c='gray')
plt.axvline(x=0, c='gray')

ax = plt.gca()
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

arrow_vector = partial(plt.arrow, width=0.05, length_includes_head=True)

arrow_vector(0, 0, 2, 1,  color='black')
arrow_vector(0, 0, 1, 2, color='black')
arrow_vector(0, 0, 2, -1, color='black')
plt.title('Linear Independence')
plt.show()
```

```python
plt.axhline(y=0, c='gray')
plt.axvline(x=0, c='gray')

ax = plt.gca()
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

arrow_vector = partial(plt.arrow, width=0.05, length_includes_head=True)

arrow_vector(0, 0, 0.1, 0.1,  color='black')
arrow_vector(0, 0, 1, 1, color='black')
plt.title('Linear Dependence')
plt.show()
```

注意：当一个向量组中存在 $0$ 向量，那么该向量组一定是线性相关的。

## 基、维数

向量空间的基（Basis）是一组具有两个特性的向量：
* 向量组线性无关
* 向量组组成平面

例如：对于三维空间 $\mathbb{R}^3$ 的一组基，$\begin{bmatrix}1\\0\\0 \end{bmatrix}, \begin{bmatrix}0\\1\\0 \end{bmatrix}, \begin{bmatrix}0\\0\\1 \end{bmatrix}$。

对于向量空间 $R^n$，它的所有基均由 $n$ 个向量组成，而 $n$ 也称为该空间的维数（Dimension）。

考虑矩阵 $A$，它的列空间维数等于它的秩，它的零空间维数等于它列数减去秩，即 $dim\ C(A) = r, dim\ N(A) = n-r$。
